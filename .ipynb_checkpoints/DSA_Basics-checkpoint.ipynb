# algorithmic complexity
It is refers to the amount of resources(time taken by algorithm or storage occupied in memory by algorithm) required to solve a problem.
## Types of complexity
* Time complexity
* ->The amount of time taken by algorithm to run on input.
* Space complexity

* using time module
* counting number of operations

## mesuring using time module

```python
import time
start=time.time()
for i in range(1,100000):
        print('*')
print('time taken',time.time()-start)
```

## 3) Order of Growth
* worst case has main emphasiz
* it deals with large input
* as input grow what happen with time.
* largest factor take in calculation
### Big(O) worst case scinario

```python
def fact(num):
    ans=1 #outer 1 operation
    while(num>1):
        ans*=num
        num -=1
    return ans
```

```python
fact(4)
```

So (1+5n) where 1 operation outside loop and 5 inside loop so 1 elminiate and last 5n that mean O(n) so time complexity is O(n) linear 
* according to input grows time also grows

Types of Big(O)
### Constant
* -->not depends on iput
* O(1)
### Linear
* -->as input grows time equally grows
* example linear search algorithm
* O(n)
### quadratic
* -->as input grows time grow by four
* O(n)^2
### logarithm
* -->as input grow time add by one
* example linear search algorithm
* O( log n)
### exponant
* as input grow time grows by 8
* O(2^n)
### n(logn)
-->example merge sort
O(n log n)

```python
def numToString(num):
    digits='0123456789'
    if num==0:
        return '0'
    else:
        result=''
        while num>0:
            result=digits[num%10]+result
            num=num//10
    return result
# input multiply so time add so log(n)
```

```python
def factorial(num):
    if num==1 or num==1:
        return 1
    elif num<0:
        return -1
    else:
        return num*factorial(num-1)
# no of function calls O(n) linear
```

```python
print(factorial(10))
```

```python
def fiboNaci(num):
    if num==1 or num==0:
        return 1
    else:
        return fiboNaci(num-1)+fiboNaci(num-2)
# exponant O(2^n) <2^n because less function calls
```

```python
print(fiboNaci(1))
print(fiboNaci(8))
# print(fiboNaci(100)) 
```

```python
def power(num):
    if num<1:
        return 0
    elif num==1:
        print(1)
        return 1
    else:
        prev=power(num//2)
        curr=prev*2
        print(curr)
        return curr
# log becuase on input increase function calls add O(n)
```

```python
power(10) #4
```

```python
power(100) #7
```

```python
power(1000)  #10
```

```python
def mod(a,b):
    if b<=0:
        return -1
    div=a//b
    return a-div*b
# constant time  O(1)
```

```python
mod(5,3)
```

```python
mod(5,4)
```

```python
mod(5,5)
```

```python
def sumDigits(num):
    sum=0
    while(num>0):
        sum+=num%10
        num//=10
    return sum
# log single loop and as input grows the time add
```

```python
sumDigits(10)
```

```python
sumDigits(123)
```
